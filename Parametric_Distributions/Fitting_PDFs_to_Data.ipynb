# Fitting a Parametric Probability Distribution to Observed Data

```python
import scipy.stats as stats # This is the statistics package
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

## Example fitting to fake data

```python
rng = np.random.default_rng()
dist = stats.nbinom
shapes = (5, 0.5)
data = dist.rvs(*shapes, size=1000, random_state=rng)

bounds = [(0, 30), (0, 1)]
res = stats.fit(dist, data, bounds)

res.params

res.plot()
plt.show()
```

## Determine the Expected Value of the PDF

```python
frozen_dist = dist(*res.params)

print(frozen_dist.expect())
print(frozen_dist.entropy())
print(frozen_dist.var())
print(frozen_dist.std())
print(frozen_dist.moment(2))
print(frozen_dist.support())
```
