# Fitting a Parametric Probability Distribution to Observed Data

```python
import scipy.stats as stats # This is the statistics package
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

np.random.seed(1)
```

## 1.) How to fit to data

```python
dist = stats.norm
mean = 0
std = 3
data = dist.rvs(loc=mean, scale=std, size=1000)

res = stats.fit(dist, data, bounds={'loc': (-1,1),\
                                    'scale': (0,6)})

print(res)
```

## Assessing the Quality of the Fit

#### Fitted PDF and Histogram

```python
res.plot()
plt.show()
```

#### Q-Q Plot

```python
z = stats.zscore(data)
stats.probplot(z, dist=dist, plot=plt, rvalue=True)
plt.title("Q-Q Plot")
plt.grid()
plt.show()
```

## 2.) Determine Properties of the Fitted PDF

```python
frozen_dist = dist(*res.params)

print("Expectation: ", frozen_dist.expect())
print("Variance: ", frozen_dist.var())
print("Standard Deviation:", frozen_dist.std())
print("1st Moment (Mean):", frozen_dist.moment(1))
print("2nd Moment (Variance):", frozen_dist.moment(2))
print("3rd Moment (Skewness):", frozen_dist.moment(3))
print("4th Moment (Kurtosis):", frozen_dist.moment(4))
```
