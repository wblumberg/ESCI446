# Demonstrating the Central Limit Theorem

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

## Creating a Normal PDF

```python
length = 100000
dataset = np.random.normal(0, 2, size=length)
plt.hist(dataset, np.arange(-10,10.1,.1))
plt.ylabel("Frequency")
plt.xlabel("Value")
plt.show()
```

## Bootstrapping any PDF to calculate the Mean

```python
means = np.zeros(10000)
stds = np.zeros(10000)
for i in range(0,len(means),1):
    dataset = np.random.lognormal(0, 2, size=length)
    means[i] = np.mean(dataset)
    stds[i] = np.std(dataset)
    
```
